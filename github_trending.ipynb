```python
import os
import datetime
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen

# function to scrape the website
def scrape():
    urls = {
        "global": {
            "daily": "https://github.com/trending?since=daily",
            "weekly": "https://github.com/trending?since=weekly",
            "monthly": "https://github.com/trending?since=monthly",
        },
        "id": {
            "daily": "https://github.com/trending?since=daily&spoken_language_code=id",
            "weekly": "https://github.com/trending?since=weekly&spoken_language_code=id",
            "monthly": "https://github.com/trending?since=monthly&spoken_language_code=id",
        },
    }

    # each url
    for key, value in urls.items():
        # each date range
        for key2, value2 in value.items():
            print("Scraping " + key + " " + key2 + "...")
            # store by key
            # mkdir date now in repositories folder
            # create filename by date now
            now = datetime.datetime.now().strftime("%Y-%m-%d")
            dirname = "repositories/" + now
            # if file not exist, create it
            if not os.path.exists(dirname):
                os.mkdir(dirname)
            # if file key not exist, create it
            if not os.path.exists(dirname + "/" + key):
                os.mkdir(dirname + "/" + key)
            if key == "global":
                f = open(dirname + "/" + key + "/" + key2 + ".json", "w")
                f.write("[")
            elif key == "id":
                f = open(dirname + "/" + key + "/" + key2 + ".json", "w")
                f.write("[")

            # create request
            req = Request(value2, headers={"User-Agent": "Mozilla/5.0"})
            client = urlopen(req)
            page_html = client.read()
            client.close()
            page_soup = soup(page_html, "html.parser")
            containers = page_soup.findAll("article", {"class": "Box-row"})
            # each container
            for container in containers:
                repository = container.h1.a.text.replace("\n", "").replace(" ", "")
                link = "https://github.com" + container.h1.a["href"]
                descriptions = container.find(
                    "p", {"class": "col-9 color-fg-muted my-1 pr-4"}
                )
                description = (
                    descriptions.text.replace("\n", "").replace('"', "'").strip()
                    if descriptions
                    else ""
                )
                languages = container.findAll(
                    "span", {"itemprop": "programmingLanguage"}
                )
                for language in languages:
                    language = language.text
                stars = container.find(
                    "a", {"class": "Link--muted d-inline-block mr-3"}
                )
                fork = (
                    stars.find_next_siblings(
                        "a", {"class": "Link--muted d-inline-block mr-3"}
                    )
                    if stars
                    else 0
                )
                stars = stars.text.replace("\n", "").replace(" ", "") if stars else "0"
                fork = fork[0].text.replace("\n", "").replace(" ", "") if fork else "0"

                print("repository: " + repository)
                print("link: " + link)
                print("description: " + description)
                print("language: " + language)
                print("stars: " + stars)
                print("fork: " + fork)
                print("\n")

                # save as json file
                f.write('{"repository":"' + repository + '",')
                f.write('"link":"' + link + '",')
                f.write('"description":"' + description + '",')
                f.write('"language":"' + language + '",')
                f.write('"stars":"' + stars + '",')
                f.write('"fork":"' + fork + '"},')
            f.write("]")


scrape()
```
